```python
HP = []
Atk = []
Def = []
SpAtk = []
SpDef = []
Spd = []
Total = []

for index in tqdm(range(df.shape[0])):
    url = 'https://pokemondb.net/pokedex/' + df['Name'].loc[index].lower()
    r = requests.get(url)
    text = r.text
    soup = BeautifulSoup(text, 'html')
    stats = soup.find_all('td')[15:43:4]
    HP.append(int(stats[0].text))
    Atk.append(int(stats[1].text))
    Def.append(int(stats[2].text))
    SpAtk.append(int(stats[3].text))
    SpDef.append(int(stats[4].text))
    Spd.append(int(stats[5].text))
    Total.append(int(stats[6].text))
```

```python
primary_type = list(df['Primary Type'])
secondary_type = list(df['Secondary Type'])
stats_dict = {'Primary Type': primary_type, 'Secondary Type': secondary_type,'HP':HP, 'Attack':Atk, 'Defense':Def, 'Special Attack':SpAtk, 'Special Defense': SpDef, 'Speed': Spd, 'Total':Total}
stats = pd.DataFrame(data = stats_dict, index = df.Name)
stats
```

```python
def get_weakness(type):
    return weaknesses[type]

def get_resistance(type):
    return resistances[type]

def get_immunity(type):
    return immunities[type]
```

```python
def get_matchup(type1, type2 = None):
    multipliers  = []
    
    if type2 != None:
        weakness1 = get_weakness(type1)
        resistance1 = get_resistance(type1)
        immunity1 = get_immunity(type1)
        
        weakness2 = get_weakness(type2)
        resistance2 = get_resistance(type2)
        immunity2 = get_immunity(type2)
        
        for index in range(df.shape[0]):
            if str(df['Secondary Type'].loc[index]) != 'nan':
                multiplier1 = 1
                multiplier2 = 1
                if df['Primary Type'].loc[index] in immunity1:
                    multiplier1 *= 0
                    
                elif df['Primary Type'].loc[index] in resistance1:
                    multiplier1 *= 0.5
                    if df ['Primary Type'].loc[index] in immunity2:
                        multiplier1 *= 0
                    elif df['Primary Type'].loc[index] in resistance2:
                        multiplier1 *= 0.5
                    elif df['Primary Type'].loc[index] in weakness2:
                        multiplier1 *= 2
                        
                elif df['Primary Type'].loc[index] in weakness1:
                    multiplier1 *= 2
                    if df ['Primary Type'].loc[index] in immunity2:
                        multiplier1 *= 0
                    elif df['Primary Type'].loc[index] in resistance2:
                        multiplier1 *= 0.5
                    elif df['Primary Type'].loc[index] in weakness2:
                        multiplier1 *= 2
                    
                elif df['Secondary Type'].loc[index] in immunity2:
                    multiplier2 *= 0
                    
                elif df['Secondary Type'].loc[index] in resistance2:
                    multiplier2 *= 0.5
                    if df ['Secondary Type'].loc[index] in immunity1:
                        multiplier2 *= 0
                    elif df['Secondary Type'].loc[index] in resistance1:
                        multiplier2 *= 0.5
                    elif df['Secondary Type'].loc[index] in weakness1:
                        multiplier2 *= 2
                        
                elif df['Secondary Type'].loc[index] in weakness2:
                    multiplier2 *= 2
                    if df ['Secondary Type'].loc[index] in immunity1:
                        multiplier2 *= 0
                    elif df['Secondary Type'].loc[index] in resistance1:
                        multiplier2 *= 0.5
                    elif df['Secondary Type'].loc[index] in weakness1:
                        multiplier2 *= 2
                        
                multipliers.append(max(multiplier1, multiplier2))
                    
            else:
                multiplier1 = 1
                if df['Primary Type'].loc[index] in immunity1:
                    multiplier1 *= 0
                    
                elif df['Primary Type'].loc[index] in resistance1:
                    multiplier1 *= 0.5
                    if df ['Primary Type'].loc[index] in immunity2:
                        multiplier1 *= 0
                    elif df['Primary Type'].loc[index] in resistance2:
                        multiplier1 *= 0.5
                    elif df['Primary Type'].loc[index] in weakness2:
                        multiplier1 *= 2
                        
                elif df['Primary Type'].loc[index] in weakness1:
                    multiplier1 *= 2
                    if df ['Primary Type'].loc[index] in immunity2:
                        multiplier1 *= 0
                    elif df['Primary Type'].loc[index] in resistance2:
                        multiplier1 *= 0.5
                    elif df['Primary Type'].loc[index] in weakness2:
                        multiplier1 *= 2
                        
                multipliers.append(multiplier1)
                
        return multipliers
    
    else:
        weakness1 = get_weakness(type1)
        resistance1 = get_resistance(type1)
        immunity1 = get_immunity(type1)
        
        
        for index in range(df.shape[0]):
            if str(df['Secondary Type'].loc[index]) != 'nan':
                multiplier1 = 1
                multiplier2 = 1
                if df['Primary Type'].loc[index] in immunity1:
                    multiplier1 *= 0
                elif df['Primary Type'].loc[index] in resistance1:
                    multiplier1 *= 0.5
                elif df['Primary Type'].loc[index] in weakness1:
                    multiplier1 *= 2
                elif df['Secondary Type'].loc[index] in immunity1:
                    multiplier2 *= 0
                elif df['Secondary Type'].loc[index] in resistance1:
                    multiplier2 *= 0.5
                elif df['Secondary Type'].loc[index] in weakness1:
                    multiplier2 *= 2
                    
                multipliers.append(max(multiplier1, multiplier2))
            
            else:
                multiplier1 = 1
                if df['Primary Type'].loc[index] in immunity1:
                    multiplier1 *= 0
                elif df['Primary Type'].loc[index] in resistance1:
                    multiplier1 *= 0.5
                elif df['Primary Type'].loc[index] in weakness1:
                    multiplier1 *= 2
                    
                multipliers.append(multiplier1)
                
        return multipliers 
```

```python
def filter_legendaries(sort_pokemon, legendary, exclude = True):
    if exclude == True:
        indices = np.array([pokemon for pokemon in sort_pokemon if pokemon not in legendary][:-6:-1])
        names = get_names(indices) 
        return names
    else:
        indices = sort_pokemon[:-6:-1]
        names = get_names(indices) 
        return names
```

```python
def get_names(indices):
    names_list = []
    for index in indices:
        names_list.append(stats.iloc[index].name)
    return names_list
```
