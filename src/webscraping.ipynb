```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
import requests


import json # to work with json file format
from bs4 import BeautifulSoup # to parse html
```

```python
import requests
url = 'https://pokemondb.net/pokedex/national'
r = requests.get(url)
```

```python
text = r.text
```

```python
soup = BeautifulSoup(text, 'html')
soup
```

```python
soup_numbers = soup.find_all('small')[0::2]
```

```python
soup_types = soup.find_all('small')[1::2]
```

```python
numbers = []
for i in range(len(soup_numbers)):
    numbers.append(soup_numbers[i].text)
```

```python
type1 = []
type2 = []
for i in range(len(soup_types)):
    if len(soup_types[i].text.split('·')) == 2:
        type1.append(soup_types[i].text.split('·')[0].strip())
        type2.append(soup_types[i].text.split('·')[1].strip())
    else:
        type1.append(soup_types[i].text.split('·')[0].strip())
        type2.append(None)
```

```python
type2
```

```python
soup_names = soup.find_all('span')
soup_names
```

```python
test = soup.find_all('span')
span = test[10].find('a')
span.get('href')
```

```python
names = []
for i in range(len(soup_names)):
    names.append(soup_names[i].text.split(' ')[1])
```

```python
names = []
for link in soup.find_all('span'):
    span = link.find('a')
    if span != None:
        name = span.get('href')
        if name != None:
            names.append(name.split('/')[2])
names = names[::2]
```

```python
names
```

```python
numS = pd.Series(numbers, name = 'ID')
nameS = pd.Series(names, name = 'Name')
typeS1 = pd.Series(type1, name = 'Primary Type')
typeS2 = pd.Series(type2, name = 'Secondary Type')
```

```python
image_links = []
for link in soup.find_all('a'):
    span = link.find('span')
    if span != None:
        image_link = span.get('data-src')
        if image_link != None:
            image_links.append(image_link)
```

```python
image_linkS = pd.Series(image_links, name = 'Image')
```

```python
df = pd.concat([numS, nameS, typeS1, typeS2, image_linkS],axis = 1)
```

```python
df.to_csv('../data/data.csv')
```

```python
extra_image_links = []
for i in range(df.shape[0]):
    url = 'https://pokemondb.net/pokedex/' + df['Name'].loc[i].lower()
    r = requests.get(url)
    text = r.text
    soup = BeautifulSoup(text, 'html')
    extra_image_link = []
    for link in soup.find_all('a'):
        span = link.find('img')
        if span != None:
            image_link = span.get('src')
            if image_link != None:
                extra_image_link.append(image_link)
    extra_image_links.append(extra_image_link)
```
